```python
import pandas as pd

data1 = pd.read_csv("Student performance in mathematics.csv")
data1
```

```python
#將 Math scorce 排名
data1_sorted = data1.sort_values('Math score')
data1_sorted
```

```python
#將 Math score 以升序排列 (由小到大)
data1_sorted = data1.sort_values('Math score')
data1_sorted
```

```python
#將 Math score 以降序排列 (由大到小)
data1_sorted = data1.sort_values('Math score', ascending=False)
data1_sorted
```

```python
#如果兩人Math score 相同，就比較Reading score 進行排名
data1_sorted = data1.sort_values(['Math score', 'Reading score'], ascending=False)
data1_sorted
```

```python
#將有準備考試的同學 數學成績相加
# 選擇 Test preparation course 欄位為 Completed 的資料
completed_data = data1[data1['Test preparation course'] == 'Completed']

# 計算 Math score 的總和
math_score_sum = completed_data['Math score'].sum()
math_score_sum
```

```python
#將沒有準備考試的同學數學成績相加
completed_data = data1[data1['Test preparation course'] == 'None']
math_score_sum = completed_data['Math score'].sum()
math_score_sum
```

```python
#將有準備考試的同學閱讀成績相加
completed_data = data1[data1['Test preparation course'] == 'Completed']
Reading_score_sum = completed_data['Reading score'].sum()
Reading_score_sum
```

```python
#將沒有準備考試的同學閱讀成績相加
completed_data = data1[data1['Test preparation course'] == 'None']
Reading_score_sum = completed_data['Reading score'].sum()
Reading_score_sum
```

```python
#將沒有準備考試的同學寫作成績相加
completed_data = data1[data1['Test preparation course'] == 'Completed']
Writing_score_sum = completed_data['Writing score'].sum()
Writing_score_sum
```

```python
#將有準備考試的同學寫作成績相加
completed_data = data1[data1['Test preparation course'] == 'None']
Writing_score_sum = completed_data['Writing score'].sum()
Writing_score_sum
```

```python
#將所有成績相加
data1['sum'] = data1['Math score'] + data1['Reading score'] + data1['Writing score']
data1['sum']
```

```python
#將所有成績相加並排名
data1['sum'] = data1['Math score'] + data1['Reading score'] + data1['Writing score']

data1['rank'] = data1['sum'].rank(ascending=False)
data1['rank']
```
