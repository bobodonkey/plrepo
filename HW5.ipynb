```python
import pandas as pd
import matplotlib.pyplot as plt

# 讀取檔案
orgdata = pd.read_csv("HW5.csv")

# 選擇所需的欄位
columns_to_compare = ["項目別", "農林漁牧業", "工業_製造業"]

# 比較個項目
comparison = orgdata[columns_to_compare]

# 選擇要比較的年齡範圍
age_ranges = ["5 ~ 19歲", "20 ~ 24歲", "25 ~ 29歲", "30 ~ 34歲", "35 ~ 39歲", "40 ~ 44歲", "45 ~ 49歲", "50 ~ 54歲", "55 ~ 59歲", "60 ~ 64歲"]

# 過濾項目別欄位中包含特定年齡範圍的數據
comparison_filtered = comparison[comparison["項目別"].str.contains("|".join(age_ranges))]

# 將年齡範圍分組，每三個一組
grouped_comparison = comparison_filtered.groupby(comparison_filtered.index // 3).sum()

# 繪製直條圖
grouped_comparison.plot(kind='bar')

# 自定義圖表標題和標籤等
plt.title('Comparison of Age Ranges in 農林漁牧業 and 工業_製造業 (Grouped)')
plt.xlabel('Age Range (Grouped)')
plt.ylabel('Values')
plt.xticks(rotation=0)  # 如果年齡範圍值過長，可以旋轉刻度標籤
plt.legend(title='Columns')

plt.show()


```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 讀取檔案
orgdata = pd.read_csv("HW5.csv")

# 選擇所需的欄位
columns_to_compare = ["項目別", "年度", "服務業_專業_科學及技術服務業"]

# 比較個項目
comparison = orgdata[columns_to_compare]

# 選擇要比較的年齡範圍
age_ranges = ["5 ~ 19歲", "20 ~ 24歲", "25 ~ 29歲", "30 ~ 34歲", "35 ~ 39歲", "40 ~ 44歲", "45 ~ 49歲", "50 ~ 54歲", "55 ~ 59歲"]

# 過濾項目別欄位中包含特定年齡範圍的數據
comparison_filtered = comparison[comparison["項目別"].str.contains("|".join(age_ranges))]

# 設定熱力圖資料
heatmap_data = comparison_filtered.pivot(index="年度", columns="項目別", values="服務業_專業_科學及技術服務業")

# 繪製熱力圖
plt.figure(figsize=(12, 8))
sns.heatmap(heatmap_data, cmap="YlGnBu")

# 自定義圖表標題和標籤等
plt.title('Comparison of "服務業_專業_科學及技術服務業" by Age Ranges and Year')
plt.xlabel('Age Ranges')
plt.ylabel('Year')

plt.show()





```
