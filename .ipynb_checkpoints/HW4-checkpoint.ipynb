```python
import pandas as pd
import matplotlib.pyplot as plt

# 讀取CSV檔案並存儲到DataFrame中
df = pd.read_csv('Train.csv')

# 選取Gender欄位為"F"和"M"的子集
subset = df[df['Gender'].isin(['F', 'M'])]

# 將Cost_of_the_Product欄位的數值除以10，並向下取整
subset['Cost_of_the_Product_Rounded'] = (subset['Cost_of_the_Product'] // 10) * 10

# 計算每個分段的頻率
frequency = subset['Cost_of_the_Product_Rounded'].value_counts().sort_index()

# 繪製直條圖
frequency.plot(kind='bar')
plt.xlabel('Cost_of_the_Product (Rounded)')
plt.ylabel('Frequency')
plt.title('Distribution of Cost_of_the_Product')
plt.xticks(rotation=45)
plt.show()



```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 讀取CSV檔案並存儲到DataFrame中
df = pd.read_csv('Train.csv')

# 選取Gender欄位為"F"和"M"的子集
subset = df[df['Gender'].isin(['F', 'M'])]

# 計算Cost_of_the_Product的總和
total_cost = subset['Cost_of_the_Product'].sum()

# 計算每個性別的Cost_of_the_Product總和
gender_costs = subset.groupby('Gender')['Cost_of_the_Product'].sum()

# 計算每個性別的占比
gender_percentages = gender_costs / total_cost * 100

# 繪製圓餅圖
labels = gender_percentages.index
sizes = gender_percentages.values
plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.axis('equal')  # 使圓餅圖保持為圓形
plt.title('Cost of the Product by Gender')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 讀取CSV檔案並存儲到DataFrame中
df = pd.read_csv('Train.csv')

# 選取Gender欄位為"F"和"M"的子集
subset = df[df['Gender'].isin(['F', 'M'])]

# 計算每個Gender和Product_importance組合的頻率
frequency = subset.groupby(['Gender', 'Product_importance']).size().unstack()

# 繪製直條圖
frequency.plot(kind='bar', stacked=True)
plt.xlabel('Gender')
plt.ylabel('Frequency')
plt.title('Distribution of Product Importance by Gender')
plt.legend(title='Product Importance')
plt.show()
```
